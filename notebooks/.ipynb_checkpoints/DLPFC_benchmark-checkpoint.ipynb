```python
import os
os.chdir("..")
import scanpy as sc
import numpy as np
import matplotlib.pyplot as plt
import torch
import scipy
import pandas as pd
import seaborn as sns
import os
import warnings
from tqdm.auto import tqdm
import torch.nn.functional as F
from collections import defaultdict
warnings.filterwarnings("ignore")
sns.set_theme("paper", style="ticks", font_scale=1.25)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
data_path = "E:/Datasets/DLPFC/151507"
adata = sc.read_visium(data_path, count_file="151507_filtered_feature_bc_matrix.h5", library_id="151507")
adata.var_names_make_unique()
annotation = pd.read_csv(os.path.join(data_path, "151507_truth.txt"), sep='\t', header=None, index_col=0)
adata.obs["Manual annotation"] = annotation.loc[adata.obs.index, 1]
# preprocessing data
sc.pp.filter_cells(adata, min_genes=20)
sc.pp.highly_variable_genes(adata, flavor="seurat_v3", n_top_genes=3000)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
adata = adata[:, adata.var['highly_variable']]
```

```python
# plot the histogram of gene expression
# if adata is sparse, convert it to dense
def characterize_exp(adata, tol=1e-3, figsize=(9.5, 4)):
    if isinstance(adata.X, scipy.sparse.csr.csr_matrix):
        exp = adata.X.toarray()
    else:
        exp = adata.X
    gene_per_spot = np.sum(exp > tol, axis=1)
    spot_per_gene = np.sum(exp > tol, axis=0)
    # get the non-zero values
    exp = exp[exp > tol]
    sparsity = 1 - exp.shape[0] / (adata.shape[0] * adata.shape[1])
    fig, axs = plt.subplots(1, 4, figsize=figsize)
    axs[0].hist(exp.flatten(), log=False)
    axs[0].set_title("Nonzero gene expression")
    axs[0].set_ylabel("log1p(count)")
    # add sparsity to the figure
    axs[0].text(0.95, 0.95, f"Sparsity: {sparsity:.2f}", transform=axs[0].transAxes, ha="right", va="top")
    # histogram of gene per spot
    axs[1].hist(gene_per_spot,  log=False)
    axs[1].set_title("Gene per spot")
    axs[1].set_ylabel("Count")
    # histogram of spot per gene
    axs[2].hist(spot_per_gene, bins=100, log=False)
    axs[2].set_title("Spot per gene")
    axs[2].set_ylabel("Count")
    # histogram of all gene expression
    exp = adata.X
    if isinstance(exp, scipy.sparse.csr.csr_matrix):
        exp = exp.toarray()
    axs[3].hist(exp.flatten(), bins=100, log=False)
    axs[3].set_title("All gene expression")
    plt.tight_layout()
    return fig


def visualize_genes(adata, adata_, gene_name):
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    sc.pl.spatial(adata, color=gene_name, ax=ax[0], show=False, title="Raw {}".format(gene_name), color_map="viridis")
    sc.pl.spatial(adata_, color=gene_name, ax=ax[1], show=False, title="Reconstructed", color_map="viridis")
    return fig
```

## Train STAGATE model

```python
import STAGATE_pyG
import scipy.sparse as sp
import torch.nn.functional as F
from tqdm.auto import tqdm
from STAGATE_pyG.STAGATE import STAGATE
from STAGATE_pyG.utils import Transfer_pytorch_Data

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = STAGATE(hidden_dims = [3000, 512, 32]).to(device)
STAGATE_pyG.Cal_Spatial_Net(adata, rad_cutoff=150)
adata.X = sp.csr_matrix(adata.X)
STAGATE_pyG.Stats_Spatial_Net(adata)
data = Transfer_pytorch_Data(adata)
data = data.to(device)
n_epochs = 1000
loss_list = []
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)
for epoch in tqdm(range(1, n_epochs+1)):
    model.train()
    optimizer.zero_grad()
    z, out = model(data.x, data.edge_index)
    loss = F.mse_loss(data.x, out) #F.nll_loss(out[data.train_mask], data.y[data.train_mask])
    loss_list.append(loss.item())
    loss.backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), 5)
    optimizer.step()
model.eval()
z, out = model(data.x, data.edge_index)
STAGATE_rep = z.to('cpu').detach().numpy()
adata.obsm["STAGATE"] = STAGATE_rep
```

##  Train VanillaCNNAE

```python
from stadiff.vqvae import VanillaCNNAE
vanilla_model = VanillaCNNAE(in_channels=adata.shape[1], latent_channels=[512, 128, 64, 30], kernel_size=[3, 3, 3, 3])
n_epochs = 1000
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
vanilla_model = vanilla_model.to(device)
optimizer = torch.optim.Adam(vanilla_model.parameters(), lr=1e-3, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=100, gamma=0.5)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
pbar = tqdm(range(1, n_epochs+1))
for epoch in range(1, n_epochs+1):
    vanilla_model.train()
    for batch_idx,  (data, slice_info) in enumerate(train_loader):
        data = data.to(device)
        optimizer.zero_grad()
        out = vanilla_model(data)
        loss = F.mse_loss(out["x_recon"], data)
        loss.backward()
        optimizer.step()
    pbar.update(1)
    scheduler.step()
    pbar.set_description(f"Epoch {epoch}, loss: {loss.item():.4f}")
# recon = exp_tensor2mat(recon, adata, adata.uns['padding_info'])
recon = out["x_recon"].squeeze(0).cpu().detach().numpy()
recon = exp_tensor2mat(recon, adata, adata.uns['padding_info'])
adata_ = adata.copy()
adata_.X = recon
latent = out["z"].squeeze(0).to('cpu').detach().numpy()
adata_.obsm["embedding"] = exp_tensor2mat(latent, adata, adata.uns['padding_info'])
```

## Train CNNVQVAE

```python
from stadiff.vqvae import CNNVQVAE
cnn_vqvae = CNNVQVAE(in_channels=3000,
                 latent_channels=[512, 128, 30],
                 num_vq_embeddings=64,
                 vq_embed_dim=30)
cnn_vqvae = cnn_vqvae.to(device)
x = exp_tensor.unsqueeze(0)
x = x.to(device)
optimizer = torch.optim.Adam(cnn_vqvae.parameters(), lr=1e-3, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=200, gamma=0.5)
tracker = defaultdict(list)
n_epochs = 500
pbar = tqdm(range(n_epochs))
for epoch in range(500):
    optimizer.zero_grad()
    out_vae = cnn_vqvae(x)
    loss = out_vae["recon_loss"] + out_vae["emb_loss"] * 0.1
    loss.backward()
    # clip grad
    torch.nn.utils.clip_grad_norm_(cnn_vqvae.parameters(), 5)
    optimizer.step()
    tracker["recon_loss"].append(out_vae["recon_loss"].item())
    tracker["emb_loss"].append(out_vae["emb_loss"].item())
    tracker["loss"].append(loss.item())
    pbar.update(1)
    pbar.set_description(f"Epoch {epoch}")
    pbar.set_postfix(loss=loss.item())
recon = out_vae["x_recon"].squeeze(0).cpu().detach().numpy()
recon = exp_tensor2mat(recon, adata, adata.uns['padding_info'])
adata_ = adata.copy()
adata_.X = recon
latent = out_vae["z"].squeeze(0).to('cpu').detach().numpy()
adata_.obsm["embedding"] = exp_tensor2mat(latent, adata, adata.uns['padding_info'])
```

## Train VQVAE model

```python
from stadiff.vqvae import VQVAE
from collections import defaultdict
import torch.optim as optim

vqvae = VQVAE(in_channels=3000, out_channels=3000, latent_channels=64, num_vq_embeddings=512, norm_num_groups=16)
vqvae = vqvae.to(device)
x = exp_tensor.unsqueeze(0)
x = x.to(device)
optimizer = optim.Adam(vqvae.parameters(), lr=1e-3)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=100, gamma=0.5)
tracker = defaultdict(list)
n_epochs = 100
pbar = tqdm(range(n_epochs))
for epoch in range(500):
    optimizer.zero_grad()
    out_vae = vqvae(x)
    loss = out_vae["recon_loss"] + out_vae["emb_loss"] * 0.1
    loss.backward()
    optimizer.step()
    tracker["recon_loss"].append(out_vae["recon_loss"].item())
    tracker["emb_loss"].append(out_vae["emb_loss"].item())
    tracker["loss"].append(loss.item())
    pbar.update(1)
    pbar.set_description(f"Epoch {epoch}")
    pbar.set_postfix(loss=loss.item())
```

## Train DDPM model

```python
os.environ["WANDB_MODE"] = "online"
os.environ["WANDB_API"] = "5cfeb02846ceabee81f586b15d4a8ba560b8cfc2"
from diffusers import DDPMScheduler
from dataclasses import dataclass
from diffusers import UNet2DModel, DDPMPipeline
from diffusers.optimization import get_cosine_schedule_with_warmup
from stadiff.pipeline import prepare_dataset
from stadiff.dataset import SpatialExpressionDataset
from stadiff.utils import spatial2index
from accelerate import Accelerator
from tqdm import tqdm_notebook as tqdm
import torch.nn.functional as F


model = UNet2DModel(
    sample_size=(96, 96),  # the target image resolution
    in_channels=3000,  # the number of input channels, 3 for RGB images
    out_channels=3000,  # the number of output channels
    layers_per_block=2,  # how many ResNet layers to use per UNet block
    block_out_channels=(32, 64, 128, 128, 128, 128),  # the number of output channels for each UNet block
    down_block_types=(
        "DownBlock2D",  # a regular ResNet downsampling block
        "DownBlock2D",
        "DownBlock2D",
        "DownBlock2D",
        "AttnDownBlock2D",  # a ResNet downsampling block with spatial self-attention
        "DownBlock2D",
    ),
    up_block_types=(
        "UpBlock2D",  # a regular ResNet upsampling block
        "AttnUpBlock2D",  # a ResNet upsampling block with spatial self-attention
        "UpBlock2D",
        "UpBlock2D",
        "UpBlock2D",
        "UpBlock2D",
    ),
)


@dataclass
class TrainingConfig:
    height = 96  # the generated image resolution
    width = 96
    train_batch_size = 1
    eval_batch_size = 1  # how many images to sample during evaluation
    num_epochs = 500
    gradient_accumulation_steps = 1
    learning_rate = 1e-3
    lr_warmup_steps = 5
    save_image_epochs = 100
    save_model_epochs = 30
    mixed_precision = "no"  # `no` for float32, `fp16` for automatic mixed precision
    output_dir = "/mnt/disk4/chzhang/projects/diffusion/output/ddpm-DLPFC-151507"  # the model name locally and on the HF Hub
    push_to_hub = False  # whether to upload the saved model to the HF Hub
    hub_private_repo = False  # whether the HF Hub repo is private
    overwrite_output_dir = True  # overwrite the old model when re-running the notebook
    seed = 0


def train_loop(config, model, noise_scheduler, optimizer, train_dataloader, lr_scheduler):
    # Initialize accelerator and tensorboard logging
    # create a folder for the logs
    accelerator = Accelerator(
        mixed_precision=config.mixed_precision,
        gradient_accumulation_steps=config.gradient_accumulation_steps,
        log_with="wandb",
        project_dir=config.output_dir,
    )
    accelerator.init_trackers("STADiff", config={}, init_kwargs={"dataset": "DLPFC-151507"})
    accelerator.print("Initialized Accelerator")
    # Prepare everything
    # There is no specific order to remember, you just need to unpack the
    # objects in the same order you gave them to the prepare method.
    model, optimizer, train_dataloader, lr_scheduler = accelerator.prepare(
        model, optimizer, train_dataloader, lr_scheduler
    )
    accelerator.print("Prepare accelerator")

    global_step = 0
    progress_bar = tqdm(total=config.num_epochs, disable=not accelerator.is_local_main_process, position=0, leave=True)
    # Now you train the model
    for epoch in range(config.num_epochs):
        progress_bar.set_description(f"Epoch {epoch + 1}")
        for step, batch in enumerate(train_dataloader):
            clean_images = batch[0]
            # Sample noise to add to the images
            noise = torch.randn(clean_images.shape).to(clean_images.device)
            bs = clean_images.shape[0]

            # Sample a random timestep for each image
            timesteps = torch.randint(
                0, noise_scheduler.config.num_train_timesteps, (bs,), device=clean_images.device
            ).long()

            # Add noise to the clean images according to the noise magnitude at each timestep
            # (this is the forward diffusion process)
            noisy_images = noise_scheduler.add_noise(clean_images, noise, timesteps)
            with accelerator.accumulate(model):
                # Predict the noise residual
                noise_pred = model(noisy_images, timesteps, return_dict=False)[0]
                loss = F.mse_loss(noise_pred, noise)
                accelerator.backward(loss)
                accelerator.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()
                lr_scheduler.step()
                optimizer.zero_grad()

            progress_bar.update(1)
            logs = {"loss": loss.detach().item(), "lr": lr_scheduler.get_last_lr()[0], "step": global_step}
            progress_bar.set_postfix(**logs)
            accelerator.log(logs, step=global_step)
            global_step += 1
        # After each epoch you optionally sample some demo images with evaluate() and save the model
        if accelerator.is_main_process:
            pipeline = DDPMPipeline(unet=accelerator.unwrap_model(model), scheduler=noise_scheduler)
            if (epoch + 1) % config.save_image_epochs == 0 or epoch == config.num_epochs - 1:
                pass
                # evaluate(config, epoch, pipeline)
            if (epoch + 1) % config.save_model_epochs == 0 or epoch == config.num_epochs - 1:
                    pipeline.save_pretrained(config.output_dir)
        # save log to local folder
        # save trakcer to local folder
    accelerator.end_training()
    return accelerator
```

```python
config = TrainingConfig()
adata = spatial2index(adata, tol=50, spot_width=137, spot_height=120)
exp_tensor, adata = prepare_dataset(adata, padding=96)
exp_tensor = torch.from_numpy(exp_tensor).float()
dataset = SpatialExpressionDataset(exp_tensor, adata.uns['padding_info'])
train_dataloader = torch.utils.data.DataLoader(
    dataset, batch_size=config.train_batch_size, shuffle=True
)
optimizer = torch.optim.AdamW(model.parameters(), lr=config.learning_rate)
lr_scheduler = get_cosine_schedule_with_warmup(
    optimizer=optimizer,
    num_warmup_steps=config.lr_warmup_steps,
    num_training_steps=(len(train_dataloader) * config.num_epochs),
)
noise_scheduler = DDPMScheduler(num_train_timesteps=1000)
args = (config, model, noise_scheduler, optimizer, train_dataloader, lr_scheduler)
```

```python
accelerator = train_loop(*args)
```

```python
from diffusers.utils import randn_tensor

min_values = []
max_values = []
gene_exp_shape = (1, model.config.in_channels, *model.config.sample_size)
sim_gene_exp = randn_tensor(gene_exp_shape, device=model.device)
progress_bar = tqdm(total=noise_scheduler.config.num_train_timesteps, disable=False, position=0, leave=True)
with torch.no_grad():
    for t in noise_scheduler.timesteps:
        model_output = model(sim_gene_exp, t).sample
        sim_gene_exp = noise_scheduler.step(model_output, t, sim_gene_exp).prev_sample
        allocated_memory = torch.cuda.memory_allocated(device=model.device)
        progress_bar.set_postfix({"allocated_memory": allocated_memory / 1e9})
        progress_bar.update(1)
        # record min and max values
        min_values.append(model_output.min().item())
        max_values.append(model_output.max().item())
```
